### **data cleaning**

```python
print(df["Item Fat Content"].unique())
```

```python
df["Item Fat Content"]=df["Item Fat Content"].replace({"LF":"Low Fat","low fat":"Low Fat","reg":"Regular"})
```

```python
print(df["Item Fat Content"].unique())
```

## **BUSINESS REQUIREMENTS**

##### **KPI'S REQUIREMENT**

```python
#Total sales
total_sales=df["Sales"].sum()
print("total sales :",total_sales)
```

```python
# average sales
average_sales=df["Sales"].mean()
print("average sales :",average_sales)
```

```python
# finding the no. of items sold
no_of_items_sold=df["Sales"].count()
print(f"no. of items sold : {no_of_items_sold}")
```

```python
# average ratings
average_ratings=df["Rating"].mean()
print(f"average ratings : {average_ratings:,.0f}")
```

### **Total Sales by Fat Content**

```python
sales_by_fat_content=df.groupby("Item Fat Content")['Sales'].sum()
print(sales_by_fat_content)
plt.pie(sales_by_fat_content, labels = sales_by_fat_content.index, autopct ='%.0f%%', startangle=90)
plt.title('Total sales by Fat content')
plt.axis('equal')
plt.show()
```

### Total sales by Item Type

```python
total_sales_by_item_type=df.groupby("Item Type")["Sales"].sum().sort_values(ascending =True)
total_sales_by_item_type
```

#### **creating pie chart **

```python
plt.figure(figsize=(10,8 ))

# 1. Remove 'labels=' inside the pie function (set it to None)
# 2. Add 'pctdistance' to push the % numbers further out
plt.pie(total_sales_by_item_type,
        labels=None,
        autopct='%.0f%%',
        startangle=90,
        pctdistance=1.1)

plt.axis('equal')
plt.title('Total sales by item type')

# 3. Create a separate legend box and move it outside
plt.legend(labels=total_sales_by_item_type.index, bbox_to_anchor=(1, 1))

plt.show()
```

## **intoducing bar chats**

```python
plt.figure(figsize=(10,8))
bars=plt.bar(total_sales_by_item_type.index,total_sales_by_item_type.values, color ='red', edgecolor='black',linewidth=2)

plt.xticks(rotation= -90)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Item Type')

for bar in bars:
  plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), s=f'{bar.get_height():.0f}', ha='center', va='bottom', fontsize =8)
plt.tight_layout()
plt.show()
```

### **FAT CONTENT BY OUTLET FOR TOTAL SALES**

```python
grouped = df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().unstack()
grouped = grouped[['Regular','Low Fat']]
ax=grouped.plot(kind='bar', stacked=False, figsize=(10,8), title='Outlet Tier by Item Fat Content')
plt.xlabel('Outlet Location Type')
plt.ylabel('Total Sales')
plt.legend(title='Item Fat Content')
plt.tight_layout()
plt.show()
```

## **Total sales by outlet performance**

```python
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()
plt.figure(figsize=(10,8))
plt.plot(sales_by_year.index,sales_by_year.values, marker='o', linestyle='-', color='b')
plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Total Sales by Outlet Establishment Year')
plt.grid(True)
plt.tight_layout()
plt.show()
```
